**Zero Shot Classification**


*   Transfer Learning
*   Unseen Labels can be determined
*   Without Specific Training
*   Helpful when lack of resources to train new model

This requires two inputs. The text on which zero-shot classification to be performed and candidate labels to which it belong






```python
classifier = pipeline(
    task = 'zero-shot-classification',
    model = 'facebook/bart-large-mnli'
)
```

```python
# Create the pipeline
classifier = pipeline(task='text-classification', model="cross-encoder/qnli-electra-base")

# Predict the output
output = classifier("Where is the capital of France?, Brittany is known for their kouign-amann.")

print(output)
```

**Text Summarization**

```python
model = 'sshleifer/distilbart-cnn-12-6'
```

```python
summarizer = pipeline(task = 'summarization', model = model)
```

```python
text = 'Artificial intelligence (AI) has rapidly evolved over the past few decades, transforming various industries and reshaping the way we live and work. From healthcare to finance, AI applications are being leveraged to improve efficiency, accuracy, and productivity. In healthcare, AI-driven diagnostic tools are enabling faster and more accurate detection of diseases, while personalized treatment plans based on AI analyses are becoming more common. In the finance sector, AI algorithms are used for fraud detection, risk management, and algorithmic trading, providing significant advantages over traditional methods. Additionally, AI is playing a crucial role in enhancing customer experiences through chatbots and recommendation systems. Despite these advancements, there are growing concerns about the ethical implications of AI, including issues related to privacy, bias, and job displacement. As AI continues to advance, it is essential to address these challenges to ensure that the benefits of AI are realized in a responsible and equitable manner.'
summary_text = summarizer(text)
```

```python
print(summary_text[0]['summary_text'])
```
